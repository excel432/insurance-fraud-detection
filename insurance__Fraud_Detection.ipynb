```python
import pandas as pd
url = "https://raw.githubusercontent.com/Talktodre-ops/techcrush-practice/refs/heads/main/insurance_claims.csv"
df = pd.read_csv(url)
df.head()
df.info()
df.describe()
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
df.isnull().sum()/len(df)*100
```

```python
df['authorities_contacted'] = df['authorities_contacted'].fillna(df['authorities_contacted'].mode()[0])
df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'])
df['incident_date'] = pd.to_datetime(df['incident_date'])
df.info()
```

```python
df = df.drop(columns = ['_c39'])
```

```python
df.info()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['fraud_reported'].value_counts()
df['fraud_reported'].value_counts(normalize=True)
```

```python
from sklearn.utils import class_weight
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['fraud_reported'] = le.fit_transform(df['fraud_reported'])
df['fraud_reported'].value_counts()
```

```python
x = df.drop(columns =['fraud_reported'])
y =df['fraud_reported']

print(x.shape)
print(y.shape)
```

```python
df_clean = x.drop(columns=['policy_number', 'policy_bind_date', 'incident_date', 'incident_location',])
print(df_clean.shape)
```

```python
categorical_cols = df_clean.select_dtypes(include = 'object').columns.tolist()
numerical_cols = df_clean.select_dtypes(include = 'number').columns.tolist()

print(categorical_cols)
print(numerical_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in categorical_cols:
  df_clean[col] = le.fit_transform(df_clean[col])
df_clean.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_clean[numerical_cols] = scaler.fit_transform(df_clean[numerical_cols])
df_clean.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_clean, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
model2 = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model2.fit(x_train_smote, y_train_smote)
y_pred2 = model2.predict(x_test)
print(classification_report(y_test, y_pred2))
print(accuracy_score(y_test, y_pred2))
```

```python
model3 = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
model3.fit(x_train_smote, y_train_smote)
y_pred3 = model3.predict(x_test)
print(classification_report(y_test, y_pred3))
print(accuracy_score(y_test, y_pred3))
```

```python
from xgboost import XGBClassifier

model4 = XGBClassifier(n_estimators=200, random_state=42, objective='binary:logistic')
model4.fit(x_train_smote, y_train_smote)
y_pred4 =model4.predict(x_test)
print(accuracy_score(y_test, y_pred4))
print(classification_report(y_test, y_pred4))
```

```python
from xgboost import XGBClassifier

model5 = XGBClassifier(n_estimators=300, max_depth=6, learning_rate=0.05, random_state=42, eval_metric='logloss')
model5.fit(x_train_smote, y_train_smote)
y_pred5 = model5.predict(x_test)
print(accuracy_score(y_test, y_pred5))
print(classification_report(y_test, y_pred5))
```

```python
y_prob = model4.predict_proba(x_test)[:, 1]
y_pred_threshold = (y_prob >=0.3).astype(int)
print(classification_report(y_test, y_pred_threshold))
```

```python
y_pred_threshold =(y_prob >=0.2).astype(int)
print(classification_report(y_test, y_pred_threshold))
```
